## Categorizando as variáveis quantitativas

```python
#Categorizando Categ_Colest
dados_cor.loc[(dados_cor['Colesterol'] <= np.quantile(dados_cor['Colesterol'], 0.25)),'Categ_Colest'] = 'menor_colesterol'
dados_cor.loc[(dados_cor['Colesterol'] > np.quantile(dados_cor['Colesterol'], 0.25)), 'Categ_Colest'] = 'colesterol_médio'
dados_cor.loc[(dados_cor['Colesterol'] > np.quantile(dados_cor['Colesterol'], 0.75)), 'Categ_Colest'] = 'maior_colesterol'
```

```python
#Categorizando BC_Max
dados_cor.loc[(dados_cor['BC_Max'] <= np.quantile(dados_cor['BC_Max'], 0.25)),'Categ_BC_Max'] = 'menor_BC_Max'
dados_cor.loc[(dados_cor['BC_Max'] > np.quantile(dados_cor['BC_Max'], 0.25)), 'Categ_BC_Max'] = 'BC_Max_médio'
dados_cor.loc[(dados_cor['BC_Max'] > np.quantile(dados_cor['BC_Max'], 0.75)), 'Categ_BC_Max'] = 'maior_BC_Max'
```

# Análise da Significância Estatística Para Análise de Correspondência Múltipla

Teste de associação qui-quadrado para avaliar a significância para pares de variáveis.

**Deve-se aplicar o teste qui-quadrado para cada combinação de pares das variáveis.** <br>

- Doença_Card X Sexo 
- Doença_Card X Tipo_Dor_Peito
- Doença_Card X Açucar_Sangue
- ...

Soma das estatísticas qui2 individuais para cada par de variável. <br>
A partir da estatística qui2 total testaremos as hipóteses:

> H0: não existe associação entre as variáveis <br>
> H1: existe associação entre as variáveis

p-valor < 0,05 rejeita H0.

C (estatística qui2) <br>
p (p-valor) <br>
dof (graus de liberdade)<br>
expected (tabela com valores esperados)<br>


**Caso alguma variável não tenha associação estatisticamente significativa com pelo menos uma outra, essa variável deve ser retirada da análise.**

```python
from scipy.stats import chi2_contingency

#Criando tabela de contingência
for column in dados_cor_select.columns:
    
    tabela_contingencia = pd.crosstab(dados_cor_select['Doença_Card'],
                                     dados_cor_select[column])
    
    # Calculando QUI2

    c, p, dof, expected = chi2_contingency(tabela_contingencia)

    print(f'Doença_Card X {column}: qui2= {c:.3f}, df= {dof}, p-value= {p:.3f}')
```

**Todas apresentam associação com pelo menos uma outra variável.**

# Análise de Correspondência Múltipla

Biblioteca Prince utiliza a Matriz de Burt para gerar as Coordenadas

```python
import prince

mca = prince.MCA(engine='auto')

#Treino do modelo
mca = mca.fit(dados_cor_select)
```

```python
#Autovalores
mca.eigenvalues_
```

```python
#Variância explicada
mca.explained_inertia_
```

```python
#Coordenadas
mca.column_coordinates(dados_cor_select)
```

## Mapa Perceptual

```python
#Plota o mapa Perceptual
mp_mca = mca.plot_coordinates(dados_cor_select,
                         ax=None,
                          figsize=(12,12),
                         show_row_points=False,
                         show_column_labels=True,
                         column_points_size = 150,
                             legend_n_cols=1)
```

```python
mp_mca = mca.plot_coordinates(dados_cor_select,
                         ax=None,
                         figsize=(12,12),
                         show_row_points=True,
                         show_column_points=False,
                         show_column_labels=False,
                         column_points_size = 150,
                             legend_n_cols=1)
```

```python
#Tratando base para plotar gráfico de dispersão
#Coordenadas das observações
mca_obs = mca.row_coordinates(dados_cor_select)
mca_obs['2'] = dados_cor_select['Doença_Card']
mca_obs.columns = ['X', 'Y', 'Doença_Card']

map_doença = {'Nao':0, 'Sim':1}
doenca = dados_cor_select['Doença_Card']

mca_obs['Doença_Card'] = doenca.map(map_doença)
mca_obs
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(x='X', 
                     y='Y', 
                     c='Doença_Card',  
                     data=mca_obs)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    title="Doença: 0-Não , 1-Sim",
                   loc='lower right')
ax.add_artist(legend1)
plt.xlabel('Component 0 16,95%')
plt.ylabel('Component 1 8,14%')

plt.show()
```
